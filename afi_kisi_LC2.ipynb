1. Berapa banyak label unique dari bike_type?

```python
df = client.query("""
SELECT COUNT(DISTINCT (bike_type)) FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

df
```

Ada 2 label unique dari kolom bike_type.

2. Apa saja nama label unique dari tipe sepeda?


```python
df = client.query("""
SELECT DISTINCT (bike_type) FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

df
```

Nama label unique dari kolom bike_type adalah classic dan electric.

3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

```python
query = """
SELECT COUNT(*) AS total_peminjaman
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE EXTRACT(YEAR FROM TIMESTAMP(start_time)) BETWEEN 2016 AND 2017
"""

df = client.query(query).to_dataframe()
df
```

Total peminjaman sepeda pada tahun 2016 dan 2017 adalah sebanyak 7993.
